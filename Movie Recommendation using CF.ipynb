```python
import numpy as np
import pandas as pd
from scipy import sparse
```

```python
#Reading movie titles
movies_title = pd.read_csv('Movielens_Dataset/movies.csv')
#Reading movie ratings
movie_ratings = pd.read_csv('Movielens_Dataset/ratings.csv')
#First marge the dataset into one single dataset as movie_ratings
#Drop genres and timestamps of the movie which are not important in using collaborative filltering
movie_ratings = pd.merge(movies_title,movie_ratings).drop(['genres','timestamp'],axis=1)
print(movie_ratings.shape)
movie_ratings.head()
```

```python
#using pivot method of pandas to create dataframe
userRatings = movie_ratings.pivot_table(index=['userId'],columns=['title'],values='rating')
userRatings.head()
print("Before: ",userRatings.shape)
#Dropping movies from dataset which have less than 15 ratings
#After dropping filling NaN with 0
userRatings = userRatings.dropna(thresh=15, axis=1).fillna(0,axis=1)
print("After: ",userRatings.shape)
userRatings.head()
```

```python
#using inbuilt correlation method to get the pearson correlation between movies
corrMatrix = userRatings.corr(method='pearson')
corrMatrix.head(100)
```

```python
#Making movie recommendation
def get_recommendation(movie_name,rating):
    #calculating the similar ratings
    similar_ratings = corrMatrix[movie_name]*(rating-2.5) 
    #subtracting the ratings by 2.5 so that the bad rated movie goes down on the list
    #sorting the values
    similar_ratings = similar_ratings.sort_values(ascending=False)
    return similar_ratings
```

```python
#creating user and entering movies and ratings for recommendation system.
user1 = [("About Time (2013)",5),("Adventureland (2009)",3),("Aliens (1986)",1),("2001: A Space Odyssey (1968)",2)]
#creating empty dataset to collect similar movies that can be recommended to this user.
similar_movies = pd.DataFrame()
for movie_title,movie_rating in user1:
    #using ignore index so that index is not created automatically on dataframe
    similar_movies = similar_movies.append(get_recommendation(movie_title,movie_rating),ignore_index = True)

similar_movies.head()
```

```python
#Summing all values row wise and sorting them in desending order to show list of movie recommednations.
print('\033[1m'+ 'Recommended Movies' + '\033[0m') #printing on bold
similar_movies.sum().sort_values(ascending=False).head(10)
#taking only 10 recommendations
```

```python
#creating another user to test the system 
user2 = [("Jumper (2008)",5),("Hellboy (2004)",4),("Toy Story 3 (2010)",2),("Kung Fu Hustle (Gong fu) (2004)",4)]
similar_movies = pd.DataFrame()
for movie_title,movie_rating in user2:
    similar_movies = similar_movies.append(get_recommendation(movie_title,movie_rating),ignore_index = True)
print('\033[1m'+ 'Recommended Movies' + '\033[0m') #printing on bold
similar_movies.head()
similar_movies.sum().sort_values(ascending=False).head(20)
```
